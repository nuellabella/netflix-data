# cleaning the dataset

```python
def cleaning_data(ndf):
    s = ndf.columns
    array_val = []
    for i in s:
        modal_val = ndf[i].mode()
        array_val.append(modal_val)
    lastest_array =np.array(array_val)
    new_array_val = [m[0] for m in lastest_array]
    modal_array_val = np.array(new_array_val)
    dict_missing =dict(zip(s, modal_array_val))
    for u,v in dict_missing.items():
        ndf[u] = ndf[u].fillna(v)
                       
                       
    return ndf
                       
train_data_clean = cleaning_data(ndf)

train_data_clean
                       
        
```

# Machine learning libraries needed

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
def model_algo(ndf,clf):
    y = ndf['classlabel']
    x = ndf.drop(['classlabel'], axis=1)
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=42)
    clf.fit(x_train, y_train)
    accuracy = metrics.accuracy_score(y_test, clf.predict(x_test))
    return accuracy
```
